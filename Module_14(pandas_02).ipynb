# **Handling Duplicates Values**

```python
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David", "Bob"],
    "City": ["New York", "London", "Paris", "New York", "Tokyo", "London"],
    "Score": [85, 90, 78, 70, 95, 90]
}

df = pd.DataFrame(data)
df
```

```python
df.duplicated()  # duplicate pele True return krbe
df.duplicated().sum() # total koita duplicate value ache
```

```python
# delelting duplicates values

df.drop_duplicates()
```

```python
df # Not deleted from the original file or not deleted permanently
```

```python
# permanent delete

df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# name er basis e duplicate deletion

df.drop_duplicates(subset = ['Name'])
```

```python
# city
df.drop_duplicates(subset=['City'])
```

```python
# kun duplicates ta rakhbo ba delete krbo tar jnno keep use krte pari


df.drop_duplicates(subset=['Name'],keep='last')
```

```python

df.drop_duplicates(subset=['Name'],keep='first')  # by default first er tai rakha hy ,,keep use charai
```

# **Apply function on df**

```python
df
```

```python
#min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()


df['Scaled Marks'] = df['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
df
```

```python
# custom built function

def grading_system(marks):
  if marks >= 260:
    return 'A+'
  elif marks>=250:
    return 'A'
  else:
    return 'A-'

df['Grade'] = df['Total Marks'].apply(grading_system)
df
```

```python
# function using for full dataframe

def marking_sys(df):
  a = df['Data Structure Marks']*2
  b = df['Python Marks']*3
  c = df['Algorithm Marks']*4

  return a+b+c

df['Exceptional Marks'] = df.apply(marking_sys,axis=1)
df
```
